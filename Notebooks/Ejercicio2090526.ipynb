```python
import pandas as pd
import numpy as np
from itertools import combinations

# 1. Cargar el archivo CSV
ruta_archivo = "https://raw.githubusercontent.com/MarioV3232/CasoPracticoProduccionIndustrial/refs/heads/main/Data/Ejercicio%202.csv"
df = pd.read_csv(ruta_archivo)

# 2. Validar dimensiones del dataset
print("Dimensiones del dataset:")
print(df.shape)

# 3. Revisar columnas disponibles
print("\nColumnas del dataset:")
print(df.columns)

# 4. Mostrar primeras filas
print("\nPrimeras filas del dataset:")
print(df.head())

# 5. Revisar tipos de datos
print("\nInformación general del dataset:")
print(df.info())

# 6. Revisar valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())

# 7. Revisar registros duplicados
print("\nCantidad de registros duplicados:")
print(df.duplicated().sum())

# 8. Estadística descriptiva general
print("\nEstadística descriptiva:")
print(df.describe())

# 9. Detección básica de valores atípicos
variables_numericas = ["Elapsed_Time", "Age"]

print("\nRevisión de valores atípicos:")
for columna in variables_numericas:
    media = df[columna].mean()
    desviacion = df[columna].std()

    limite_inferior = media - 2 * desviacion
    limite_superior = media + 2 * desviacion

    outliers = df[
        (df[columna] < limite_inferior) |
        (df[columna] > limite_superior)
    ]

    print("Variable:", columna)
    print("Media:", round(media, 2))
    print("Desviación estándar:", round(desviacion, 2))
    print("Límite inferior:", round(limite_inferior, 2))
    print("Límite superior:", round(limite_superior, 2))
    print("Cantidad de valores atípicos:", len(outliers))
    print("-" * 40)

# 10. Revisión de variables categóricas
print("\nDistribución de Time_in_Community:")
print(df["Time_in_Community"].value_counts())

print("\nDistribución de Gender:")
print(df["Gender"].value_counts())

print("\nDistribución de Working:")
print(df["Working"].value_counts())

# 11. Selección de columnas para reglas de asociación
columnas_asociacion = [
    "Family",
    "Hobbies",
    "Social_Club",
    "Political",
    "Professional",
    "Religious",
    "Support_Group"
]

df_asociacion = df[columnas_asociacion].copy()

print("\nPrimeras filas del dataset de asociación:")
print(df_asociacion.head())

# 12. Validación de valores binarios
print("\nValores únicos en columnas de asociación:")
for columna in columnas_asociacion:
    print(columna, df_asociacion[columna].unique())

# 13. Conversión de 0/1 a booleanos
df_asociacion_bool = df_asociacion.astype(bool)

print("\nTipos de datos después de convertir a booleanos:")
print(df_asociacion_bool.dtypes)

print("\nPrimeras filas del dataset booleano:")
print(df_asociacion_bool.head())

# 14. Frecuencia absoluta por tipo de organización
frecuencias = df_asociacion.sum().sort_values(ascending=False)

print("\nFrecuencia de pertenencia por organización:")
print(frecuencias)

# 15. Porcentaje de participación por tipo de organización
porcentajes = (df_asociacion.mean() * 100).sort_values(ascending=False)

print("\nPorcentaje de participación por organización:")
print(porcentajes.round(2))

# 16. Revisión preliminar de combinaciones frecuentes
print("\nCombinaciones con soporte mayor al 10%:")
for grupo_a, grupo_b in combinations(columnas_asociacion, 2):
    soporte = (df_asociacion_bool[grupo_a] & df_asociacion_bool[grupo_b]).mean()

    if soporte > 0.10:
        print(grupo_a, "+", grupo_b, "=", round(soporte, 4))

# 17. Guardar dataset preparado
df_asociacion_bool.to_csv("Ejercicio_2_preparado_asociacion.csv", index=False)

print("\nDataset preparado correctamente para reglas de asociación.")
```
